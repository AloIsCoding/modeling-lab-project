# 1. Cost functions

Lets focus on linear regression of the form 

$\mathbf{y} \approx f(\mathbf{X}) = \mathbf{X}\mathbf{w_1} + \mathbf{w_0}.$


#### 1.1 What are the rows of $\mathbf{X}$?

The rows of $\mathbf{X}$ correspond to the data points (sample / observation). A row contains the values of
all features for that sample. It could list properties, like genre, length, etc.

#### 1.2 What are the columns of $\mathbf{X}$?

The columns of $\mathbf{X}$ are the features or variables describing the observations. A column lists the values of that feature cross all samples and each column describes a specific feature across all the data points. 

Often, we write the equation above as

$\mathbf{y} \approx \mathbf{\tilde{X}}\mathbf{w}$

#### 1.3 How does $\mathbf{\tilde{X}}$ look like in this case (i.e., how does the shape of the matrix change compared to $\mathbf{X}$)?

The $\mathbf{w_0}$ has been integrated into the $\mathbf{\tilde{X}}$, so the latter contains one more column, whereas in the expression with $\mathbf{X}$ and $\mathbf{w_0}$ the bias is separate from $\mathbf{X}$.
The expression with $\mathbf{\tilde{X}}$ must facilitate further calculations.

For machine learning, we need a cost function. Two common choices are the mean-squared error (MSE, $\mathcal{L}_2$), and the mean-absolute error (MAE, $\mathcal{L}_1$)

\begin{align}
    \mathcal{L}_2 &=& \frac{1}{N} \sum_{i=1}^N \left(y_i - f(x_i) \right)^2 \\
    \mathcal{L}_1 &=& \frac{1}{N} \sum_{i=1}^N \left|y_i - f(x_i) \right| 
\end{align}

#### 1.4 In the Jupyter notebook, write a Python function that computes these two cost functions given an error term $\boldsymbol{\epsilon} = \mathbf{y} - \mathbf{\tilde{X}}\mathbf{w}$

````python
def mean_squared_error(error_vector):
    mean_squared_error = np.mean((error_vector)**2)
    return mean_squared_error
````

````python
def mean_absolute_error(error_vector):
    mean_absolute_error = np.mean(abs(error_vector))
    return mean_absolute_error
````

# 2. Regularization

Assume that the columns of $\mathbf{X}$ are linearly independent.
As a refresher of linear algebra, recall when the linear system $\mathbf{X}\mathbf{w} = \mathbf{y}$ has

#### 2.1 One unique solution

The number of equations $m$ is equal to the number of unknown $n$ for a $m \times n$ matrix. It is a __square matrix__ with linearly independent rows.

#### 2.2 No solution

The matrix is __not consistent__. There is no linear combination of X columns that can generate y.

#### 2.3 An infinite number of solutions

The system is __undetermined__. This means that the number of unknows $n$ is greater than the number of equations $m$, but the system is consistent.

#### 2.6 Differentiate above formula step by step and show what will we have?

We start from the least-squares cost function:

$$
f(w) = \|y - \tilde{X}w\|_2^2 = (y - \tilde{X}w)^\top (y - \tilde{X}w)
$$

**Step-by-step expansion:**

$$
\begin{aligned}
f(w)
&= y^\top y - y^\top \tilde{X} w - w^\top \tilde{X}^\top y + w^\top \tilde{X}^\top \tilde{X} w \\
&= y^\top y - 2w^\top \tilde{X}^\top y + w^\top \tilde{X}^\top \tilde{X} w
\end{aligned}
$$

**Derivative with respect to $w$:**

Using $ \frac{\partial (w^\top A w)}{\partial w} = 2Aw $ (for symmetric $A$) and $ \frac{\partial (b^\top w)}{\partial w} = b $,

$$
\frac{\partial f}{\partial w} = -2\tilde{X}^\top y + 2\tilde{X}^\top \tilde{X}w
$$

That is the gradient of the least-squares cost.


#### 2.7 if we want $\| y - \tilde{X} w \|_2^2$ to be minimum, what should the derivative be equal to?

To minimize $f(w)$, we set its derivative to zero:

$$
\frac{\partial f}{\partial w} = 0
$$

This gives:

$$
-2\tilde{X}^\top y + 2\tilde{X}^\top \tilde{X}w = 0
$$

Simplifying:

$$
\boxed{\tilde{X}^\top \tilde{X}w = \tilde{X}^\top y}
$$

These are known as the **normal equations**.  
If $\tilde{X}^\top \tilde{X}$ is invertible, the closed-form solution is:

$$
\boxed{w = (\tilde{X}^\top \tilde{X})^{-1}\tilde{X}^\top y}
$$


#### 2.8 What is the Hat matrix and what does its diagonal values correspond to?

The **Hat matrix** maps observed targets $y$ to predicted values $\hat{y}$:

$$
\boxed{H = \tilde{X}(\tilde{X}^\top \tilde{X})^{-1}\tilde{X}^\top}
$$

and

$$
\hat{y} = H y
$$

**Properties:**
- $H$ is **symmetric**: $H^\top = H$
- $H$ is **idempotent**: $H^2 = H$
- It represents an **orthogonal projection** of $y$ onto the column space of $\tilde{X}$

**Diagonal elements $h_{ii}$:**

$$
h_{ii} = x_i (\tilde{X}^\top \tilde{X})^{-1} x_i^\top
$$

where $x_i$ is the $i^{th}$ row of $\tilde{X}$.

Each $h_{ii}$ measures the **leverage** of observation $i$:
- High $h_{ii}$: observation has unusual $x$-values and strong influence on its fitted value.
- The average leverage is $p/n$, where $p$ = number of model parameters.
- The sum of leverages equals the number of parameters: $\text{trace}(H) = p$.


#### 2.9 What is Willson‚Äôs plot and how does it help in outlier detection?

The **Williams plot** (also called *Residuals vs Leverage* plot) is a diagnostic tool that plots:

- **x-axis:** the leverage $h_{ii}$
- **y-axis:** the standardized or studentized residuals $r_i$

**Interpretation:**
- Points with large residuals ($|r_i| > 2$ or $3$) ‚Üí potential *outliers in $y$* (response outliers).
- Points with high leverage ($h_{ii} > 2p/n$) ‚Üí potential *outliers in $x$* (influential points in feature space).
- Points with both high leverage and large residuals ‚Üí *influential observations* that may distort the model fit.

**How it helps:**
- Identifies data points that strongly affect the fitted regression line.
- Often used with **Cook‚Äôs distance** for quantitative influence measurement.

A Williams plot separates:
- normal points (low leverage, small residuals),
- high-leverage points,
- and true outliers (large residuals and leverage).

#### 2.10 What happens if some columns are linearly dependent? What is the connection to feature selection?

If some columns of $\tilde{X}$ are **linearly dependent**, then:

$$
\tilde{X}^\top \tilde{X} \text{ is singular (non-invertible).}
$$

**Consequences:**
- The normal equations $ \tilde{X}^\top \tilde{X}w = \tilde{X}^\top y $ have **no unique solution**.
- Parameter estimates become **unstable** and highly sensitive to noise (a problem known as *multicollinearity*).

**Connections to feature selection:**
- Linearly dependent (or highly correlated) features carry **redundant information**.
- Feature selection removes or combines such features to reduce redundancy and improve model stability.
- Alternatively, **regularization** (e.g., Ridge regression) ensures invertibility even with collinear features.

In summary:  
Feature selection or regularization helps control linear dependence, improving interpretability and numerical stability.

#### 2.11 What will be the new cost function after adding the regularization term?

After adding the L2 regularization term, the cost function becomes:

$$
\boxed{f_\lambda(w) = \|y - \tilde{X}w\|_2^2 + \lambda \|w\|_2^2}
$$

or equivalently,

$$
f_\lambda(w) = (y - \tilde{X}w)^\top (y - \tilde{X}w) + \lambda w^\top w
$$

**Interpretation:**
- The first term measures the data fit (residual sum of squares).
- The second term penalizes large weights, controlling model complexity.
- $\lambda \ge 0 $ :
  - $ \lambda = 0 $ : ordinary least squares (no penalty).  
  - $ \lambda > 0 $: smoother model, smaller coefficients.


#### 2.12 What will be the new ùë§ when we differentiate the new cost function and set it to zero.

Start from the regularized cost:

$$
f_\lambda(w) = (y - \tilde{X}w)^\top (y - \tilde{X}w) + \lambda w^\top w
$$

**Derivative with respect to $w$:**

$$
\frac{\partial f_\lambda}{\partial w} = -2\tilde{X}^\top y + 2\tilde{X}^\top \tilde{X}w + 2\lambda w
$$

Set the derivative equal to zero:

$$
-2\tilde{X}^\top y + 2(\tilde{X}^\top \tilde{X} + \lambda I)w = 0
$$

Simplify:

$$
(\tilde{X}^\top \tilde{X} + \lambda I)w = \tilde{X}^\top y
$$

**Closed-form solution:**

$$
\boxed{w_\lambda = (\tilde{X}^\top \tilde{X} + \lambda I)^{-1}\tilde{X}^\top y}
$$

where **$I$** is the *identity matrix* of size equal to the number of features (i.e. the same dimension as $\tilde{X}^\top \tilde{X}$).

**Notes:**
- For $\lambda > 0$, the matrix $(\tilde{X}^\top \tilde{X} + \lambda I)$ is always invertible (positive definite).
- Regularization guarantees stability of the solution and reduces overfitting.



#### 2.13 Prove that the part of Hat matrix where we want to take the inverse from is always reversible after we introduce the regularization term.

In the case of ordinary least squares it was determined that the hat matrix is of the following form:

$X(X^{T}X)^{-1}X^{T}$

in the case where the column of X are not linearly independant, $X^{t}X$ has no inverse.

For regularized least square, the hat matrix is written as :

$X(X^{T}X+\lambda I)^{-1}X^{T}$

Let's write the quadratic form of $v^{T}Hv$ with $v$ a non zero vector, $H$ the hat matrix:

$v^{T}(X^{T}X+\lambda I)v = v^{T}X^{T}Xv + \lambda v^{T}v = ||Xv||^{2}_{2} + \lambda ||v||^{2}_{2}$

and $||Xv||^{2}_{2}>0$ and $\lambda ||v||^{2}_{2}>0$ for all $\lambda >0$, by comparing the structure one can conclude that $X^{T}X+\lambda I$ is a positive definite matrix, thus is invertible.

#### 2.15 Plot the approximation to the function for different order polynomials ($N \in \{1, 2, 16\}$) and with different regularization strength ($\lambda \in \{0, 10^{-3}, 10^{-2}, 1\}$). What do you observe 

````python
def true_function(X):
    return np.cos(1.5 * np.pi * X)
````

````python
X_test = np.linspace(0, 1, 100) # some grid for us on the x axis
````

````python
n_samples = 10 # the number of points we will sample from true_function
degrees = [1, 2, 16] # the polynomial degrees we will test
lambdas=[0,1e-3,1e-2,1]
X = np.sort(np.random.rand(n_samples))
y = true_function(X) + np.random.randn(n_samples) * 0.1 # add some scaled random noise
````

````python
plt.scatter(X, y, label='noisy samples')
plt.plot(X_test, true_function(X_test), c='r', label='true function')
plt.legend()
````

The following code will fit a polynomial regression, you need to fill the degree

````python
for i, d in enumerate(degrees):
    for j, lam in enumerate(lambdas):
        if lam==0:# corresponds to the case of an ordinary least square
            model=Pipeline([('polynomial', PolynomialFeatures(degree=d, include_bias=False)),("linear_regression", LinearRegression())])
        else:
            model = Pipeline([("polynomial", PolynomialFeatures(degree=d,include_bias=False)),("ridge",Ridge(alpha=lam))])
        
        model.fit(X[:, None], y)
````

To plot the result, you can use the following code

````python
plt.plot(X_test, model.predict(X_test[:, np.newaxis]), label="model of degree(1 to 16)")
plt.scatter(X, y, label='noisy samples')
plt.plot(X_test, true_function(X_test), c='r', label='true function')
plt.legend()
````

Different order of polynomials and the effect of the regularization strength was tested as seen bellow using the structures given above. 

````python
degrees = [1, 2, 16]
lambdas = [0, 1e-3, 1e-2, 1]

plt.figure(figsize=(10, 6))  # make the figure a bit larger

for i, d in enumerate(degrees):
    for j, lam in enumerate(lambdas):
        if lam == 0:  # ordinary least squares
            model = Pipeline([
                ('polynomial', PolynomialFeatures(degree=d, include_bias=False)),
                ("linear_regression", LinearRegression())
            ])
        else:
            model = Pipeline([
                ("polynomial", PolynomialFeatures(degree=d, include_bias=False)),
                ("ridge", Ridge(alpha=lam))
            ])
        
        model.fit(X[:, None], y)
        y_prediction = model.predict(X_test[:, None])
        plt.plot(X_test, y_prediction, lw=1.2, label=f'Œª={lam:g}')  # thinner lines + simpler label
    plt.scatter(X, y, color='black', label='noisy samples')
    plt.plot(X_test, true_function(X_test), 'r--', lw=2, label='true function')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=1)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f'model for degree={d}')
    plt.tight_layout()
    plt.show()
````

One can observe that the more degrees there are the closer to the true function f(x) will get, one can also notice that $\lambda$ is relevent as the bigger it is the flatter f(x) becomes which in some case is better to simulate the true function and in other cases less appropriate. For instanc eone can see above that for smaller degrees a smaller $\lambda$ is better however, for higher degrees the lowest lambda fails to better simulate the true function. Such an effect is due to the fact that $\lambda$ represent a constraint on big coefficient, the higher its value the bigger the constraint.

#### 2.17 Why do we need a test set in machine learning?

A test set in maching learning is crucial to verify the ability of a trained model to treat data that it has never used before. test sets are set of data that are unbiased regarding the model we are developping.

#### 2.18 If we need to optimize hyperparameters, do we use the test set to select the best hyperparameters? (1 point)

It is better not to use the test set to determine the best hyperparameter as this would mean running the test set over and over which might indirectly lead to the model "learning" from the test set, thus the test set will not remain unbiased and will loose its purpose. It is better to use a set dedicated to the selection of a better parameter and keep the test set for the final steps of developping a working model.
